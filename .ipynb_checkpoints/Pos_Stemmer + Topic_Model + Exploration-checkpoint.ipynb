```python
import pandas as pd
import numpy as np
import seaborn as sn 
import os, glob

# Load datasets
os.chdir('/Users/daniellee/Desktop/Kaggle/jobs_code_demo/data/stackoverflow_data/')
train = pd.read_csv('cleaned/df_topics.csv'); train.head(3)
test = pd.read_csv('test_set/test.csv'); train.head(3)
```

```python
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
from nltk import pos_tag as pt
from nltk.stem import WordNetLemmatizer
import re, string

raw_df = pd.concat(map(pd.read_csv, glob.glob("*.csv")))[['title', 'content']]

uri_re = r'(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:\'".,<>?«»“”‘’]))'
def stripTagsAndUris(x):
    """ Strip out HTML tags """
    
    if x:
        # BeautifulSoup on content
        soup = BeautifulSoup(x, "html.parser")
        # Stripping all <code> tags with their content if any
        if soup.code:
            soup.code.decompose()
        # Get all the text out of the html
        text =  soup.get_text()
        # Returning text stripping out all uris
        return re.sub(uri_re, "", text)
    else:
        return ""
    
def posPlucker(x):
    """ Retrieves words with JJ, NN, NNP, NNPS, and VB POS tags """
    
    return ' '.join([word[0] for word in pt(x.split()) if word[1] in ['NN','NNP','NNPS','NNS','JJ','VB','VBN']])
    
    
punctuation = string.punctuation.replace('-','')
def removePunctuation(x):
    """ Removes punctuation marks """
    
    # Lowercasing all words
    x = x.lower()
    # Removing non ASCII chars
    x = re.sub(r'[^\x00-\x7f]',r' ',x)
    # Removing (replacing with empty spaces actually) all the punctuations
    return re.sub("["+punctuation+"]", " ", x)

wnl = WordNetLemmatizer()
def lemmatizer(x):
    """ Removes 's' at the end of a word """
    
    # Removes the letter s
    return ' '.join(map(wnl.lemmatize, x.split()))

stops = set(stopwords.words("english"))
def removeStopwords(x):
    """ Removes English stopwords """
    
    # Removing all the stopwords
    filtered_words = [word for word in x.split() if word not in stops]
    return " ".join(filtered_words)

# Create new columns
for cleaner in [stripTagsAndUris, posPlucker, removePunctuation, lemmatizer, removeStopwords]:
    raw_df['title'] = raw_df['title'].map(cleaner)
    raw_df['content'] = raw_df['content'].map(cleaner)
    test['title'] = test['title'].map(cleaner)
    test['content'] = test['content'].map(cleaner)
```

```python
train['title_nouns'] = raw_df['title'].values
train['content_nouns'] = raw_df['content'].values
```

```python
train['nouns_size'] = train.title_nouns.apply(lambda df: len(df.split()))
```

```python
# Output the dataframe for re-use 

train.to_csv('cleaned/topic_model_df_train.csv')
train.to_csv('cleaned/topic_model_df_test.csv')
```

## POV

```python
from ast import literal_eval

train['tags'] = [literal_eval(i) for i in train.tags] 
```

```python
from itertools import chain

allTags = pd.Series(list(chain.from_iterable([tags for tags in train.tags])))
uniqueTags = pd.Series(allTags.unique())

pov = pd.Series([pair[1] for pair in pt(uniqueTags)])
wordPosDF = pd.concat([uniqueTags, pov], axis=1); wordPosDF.columns = ['tags','pov']; wordPosDF.T.head()
```

```python
wordPosDF['pov'].value_counts(ascending=False)
```
